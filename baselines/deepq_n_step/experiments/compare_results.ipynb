```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
def load_progress_data(data_path):
    data_frames = []
    for sub_dir in os.listdir(data_path):
        data_frames.append(pd.read_csv(os.path.join(data_path, sub_dir, 'progress.csv'),  sep=','))
    return data_frames
```

```python
def plot_data(data_frames = [], xlabel='steps', ylabel='mean 100 episode reward', legend=[], title=''):
    fig, axes = plt.subplots(1,1)
    for data in data_frames:
        axes.plot(data[xlabel], data[ylabel])
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.legend(legend)
    axes.set_title(title)
```

# Plot Mean and Std

```python
def plot_data_group(data_groups = [], legend=[], title='', xlabel='steps', ylabel='episode reward'):
    fig, axes = plt.subplots(1,1)
    max_steps = 100000
    std_proportion = 1#1/4
    for group in data_groups:
        aggregate = np.zeros((len(group), max_steps))
        for data_i in range(len(group)):
            aggregate[data_i, :] = np.interp(np.arange(0, max_steps), group[data_i]['steps'], group[data_i]['mean 100 episode reward'])
        data_mean = np.mean(aggregate, axis=0)
        data_std = np.std(aggregate, axis=0)
        axes.plot(data_mean)
        axes.fill_between(np.arange(0, max_steps), data_mean+data_std*std_proportion, data_mean-data_std*std_proportion, alpha=0.15)
    axes.legend(legend)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.set_title(title)
    plt.savefig('./figures/comparison_{}.jpg'.format(title), dpi=300)
```
